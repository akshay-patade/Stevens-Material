```python
### Q1 Print(10 pts)

### 1. Define a string (Hello Python!), print it.

hello = "Hello Python!"
print(hello)

### 2. Define a string (Let’s write code.), print it.

code = "Let’s write code."
print(code)

### 3. Define a string and print:

python = """Python is a high-level, general-purpose
programming language. Its design
philosophy emphasizes code readability
with the use of significant indentation."""

print(python)

### Q2 Operations(15 pts)

### Define a = 100, b = 9, c = 23, calculate the results of the following problems.
a,b,c = 100, 9, 23

### 1. (2pts) Assign the result of (a + c)b to d, print d out.
d = (a + c) * b
print(d)

### 2. (2pts) Print the result of c/b.
print(c / b)

### 3.(2pts) Print the integer part of d/b. (Hint: modulo operations, calculate quotients)
print(d // b)

### 4.(2pts) Print the remainder part of d/b.
print(d % b)

### 5. (2pts) Print the result of a^2 (a squared)
print(a * a)

### 6. (2pts) Print the result of a^2 (a squared)
print(b != c)

### 7. (2pts) Determining if a is less or equal to b and print the False using logic operations.
print(a <= b)


## Q3 List Practice(20 pts)
#### 1. (4pts) Define a 10-element list with consecutive integers starting from 0. Name it as list a. Hint: Use range()
a = []
for i in range(10):
    a.append((i))

print(a)

#### 2. (4pts) Using extend to concatenate another list ([10, 11]) to list a, conclude the difference between .append() and .extend()
a.extend(([10,11]))
print(a)

###Append adds a single element to the list whereas extend adds multiple individual elements to the list
#### 3. (4pts) Insert a string (’FE520’) to the second place of list a, and delete it after that.
print("Adding FE520 to the second place and printing it")

#The index position starts with index 0 so the second place would be index 1
a.insert(1, 'FE520')
print(a)

print("Removing FE520")
a.pop(1)

print(a)

### 4. (4pts) Create a sub-list from list a which doesn’t contain the last element of it. Name it list b.
print("List b")
b = a[0:11]
print(b)

#### 5. (4pts) Slice list b from the end to the 3rd element. Override the list b with this sub-list.
print("Slicing operation")
b = b[2:]
print(b)

### 6.  (5pts) Create a copy of list b, name it list c. Concatenate list b and list c to form list d.
print("Printing d")

c = b.copy()

d = b + c

print(d)

print("Reverse the contents of d and print it")
d.reverse()
print(d)

## Q4 Dictionary Practice(20 pts)
## 1. (2pts) Define a string of ‘python is an interpreted dynamic programming language’
str1 = "python is an interpreted dynamic programming language"
print(str1)

## 2. (3pts) Create a list (list A) of single-character strings out of the above string in 1 (e.g., ’hello’->[’h’, ’e’, ’l’, ’l’, ’o’]).
a = list(str1)
print(a)

#### 3. (8pts) Write a loop to count the number of each unique character (ignore spaces) into dictionary, where your keys are characters in the list A, and  value is the count corresponding to each character. Your result should look like : {’h’: 1, ’e’: 1, ’l’: 2, ’o’: 1}.
dic = {}
for i in range(0, len(a)):
    
    if(a[i] == " "):
        continue
    
    if(a[i] in dic):
        dic[a[i]] +=  1
    
    else:
        dic[a[i]] = 1

print(dic)

#### 4. (7pts) Print the characters which shows up more than 1 time. (Hint: use loop and if statement).
for key in dic:
    if(dic[key] > 1):
        print("%s:%d" %(key, dic[key]))

#Q5 Loop Practice: Sum(15 pts)
#Write a loop for calculate the average of a list
a = [2, 3, 7, 10, 33, 12]

sum_list = 0
for i in range(len(a)):
    sum_list += a[i]
    
avg = sum_list / len(a)

print(avg)

## Q6 Control Flow Practice: Element-wise logical operations (20 pts)
"""1. (10 pts) Given two same-length lists of bool-type variables (True/False), you need
to implement the element-wise and, or, and not operations. Element-wise means the
operations are applied to each pair of two element, and the results form another list.
In the following sample code, the first element in result and is False because x[0]
and y[0] equals False (True and False == False). Complete the following code to
obtain the three lists. """

x = [True, True, False, False, True, False, True]
y = [False, True, True, True, False, False, True]

result_and = []
result_or = []
result_notx = []

for a, b in zip(x, y):
    result_and.append(a and b)
    result_or.append(a or b)
    result_notx.append(not a)

print(result_and)
print(result_or)
print(result_notx)

""" 2. (10 pts) Implement another operation: given a list of bool-type variables. If True
takes 40% or more of the total number of element of this list, give True as the result,
otherwise give False."""

from random import randint
import random
list_test = [bool(randint(0,1)) for _ in range(40)]

print("Newly generated list")

print(list_test)

count = 0;
for i in range(len(list_test)):
    if(list_test[i]):
        count += 1

print("Printing if the total number of true in list_test is greater than or equal to 40%")

print(count >= len(list_test))
    

""" Random int is one of the method from the random class to generate random numbers from
a particular range. In the above example we used randint(0,1). So our program will generate only two random 
integers 0 and 1 and it will store it in the list. An alternate way to store the random numbers in the list is to use
random.sample method
 """
  
import numpy as np
import random

random_list = np.random.choice([True, False], size = 40)
print(random_list)
 
        
```
