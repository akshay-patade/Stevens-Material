# Implement Apply( ) Method (10 pts)

```python
# now try to label the column annual income
# the criterion is like the following:
# if the annual income is larger than or equal to 800000, the label is "excellent"
# if the annual income is between 600000 and 800000 (include 600000 but not include 800000), the label is good
# store the label created in a new column called "classify annual income"

########

def classify_annual_income(income):
    if income >= 800000:
        return 'excellent'
    elif 600000 <= income < 800000:
        return 'good'
    else:
        return 'ok'

########
```

```python
data["classify annual income"] = data["amt_income_total"].apply(classify_annual_income)
data.loc[data["classify annual income"] == "excellent"]
```

```python
# in this practise, we are trying to use binary label to represent column code_gender
# if the gender is male, we label it as 1
# if the gender is female, we label it as 0
# we make this converting on the original column


########

map_gender = {'M' : 1, 'F': 0}

########
```

```python
data["code_gender"] = data["code_gender"].apply(lambda x: map_gender[x])
data
```
